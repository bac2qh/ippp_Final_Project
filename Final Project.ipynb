```python
# so this is how we use api

import pandas as pd
import json
import requests
from urllib.parse import quote

soql = "https://data.cityofchicago.org/resource/6zsd-86xi.json?$query="
soql += quote("SELECT community_area AS Community, count(*) AS Count ")
soql += quote("WHERE 2007 < year AND year < 2013 ")
soql += quote("GROUP BY Community LIMIT 10000")

resp = requests.get(soql).json()

# notice the 'count' function in line 2
# you should be able to play around it to get whatever you want
# I think we can stick with count for now

result = pd.DataFrame(resp)
# To check if this data is valid,
# I summed up all the count and happily 
# found out that it equals len(crime_0812_1)

result['Count'].astype(int).sum()
```
